```python
import numpy as np 
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# graphique de clasification

iris = load_iris()
x= iris.data
y=iris.target

names = list (iris.target_names)
print (names)
print (f" x contient {x.shape[0]} exemples et {x.shape[1]} variables")
print (f" il y'a {np.unique(y).size} calsses" )

```

```python
n = x.shape[1] 
plt.figure (figsize=(12,8))
for i in range(n) :
    plt.subplot( n // 2, n//2 , i+1)
    print (i)
    plt.scatter( x[ :, 0], x[:, i], c=y)
    plt.xlabel ( "0")
    plt.ylabel( i)
    plt.colorbar(ticks=list(np.unique(y)))

plt.show()
    
```

```python
# graphique 3D
from mpl_toolkits.mplot3d import Axes3D

ax = plt.axes(projection ='3d')
ax.scatter( x[ :, 0],x[:, 1], x[:,3], c=y)
```

```python
import numpy as np 
import matplotlib.pyplot as plt
f = lambda x, y : np.sin(x) + np.cos(x+y) 
X = np.linspace(0, 5, 50)
Y = np.linspace(0, 5, 50)
X, Y = np.meshgrid(X, Y)
Z = f(X, Y)
ax = plt.axes(projection = '3d')
ax.plot_surface( X, Y, Z, cmap = 'plasma')
plt.show()
```

```python
import numpy as np 
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

iris = load_iris()
x= iris.data
y=iris.target

plt.hist( x[ : , 0], bins= 30, color = "green", label = 'langeur du spale')
plt.hist(x[ : , 1], bins =20, color= 'yellow', label = 'largeur du spale')
plt.title("distribution des longueurs et la largeur de spales, pour les diffrentes fleurs dans le jeu de donnes Iris.")
plt.legend()
plt.show()
```

```python
plt.hist2d(x[:,0], x[:,1], cmap='Blues')
plt.xlabel('longueur spal')
plt.ylabel('largeur spal')
plt.colorbar()
```
