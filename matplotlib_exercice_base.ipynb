```python
# Créez une fonction "graphique" qui permet de tracer sur une seule et meme figure une série de graphiques
# issue d'un dictionnaire contenant plusieurs datasets 

# Voici le dataset utilisé

import numpy as np 
import matplotlib.pyplot as plt
dataset = {f"experience{i}": np.random.randn(100) for i in range(6)}

dataset
```

```python
def graphique (dataset) : 
    n = len(dataset)
    plt.figure(figsize=(12, 20))

    for k, i in zip(dataset.keys(), range(1, n+1)): 
        plt.subplot ( n , 1, i)
        plt.plot(dataset[k])
        plt.title(k)

plt.show()
```

```python
graphique(dataset)
```
